# Data Manupulation

```python
Data = Data.astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
for column in Data.columns:
    Data[column] = LE.fit_transform(Data[column])
```

```python
Data.head()
```

# Storing Feature matrix in X and target in vector y 

```python
X = Data.drop('class',axis=1)
y = Data['class']
```

```python
X
```

```python
y
```

# Applying PCA 

```python
# Reducing the no.of variables 
from sklearn.decomposition import PCA
pca1 = PCA(n_components=7)
PCA = pca1.fit_transform(X)
```

```python
PCA
```

```python
pca1.explained_variance_ratio_
```

```python
sum(pca1.explained_variance_ratio_)
```

# Splitting the data set into traing and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(PCA, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
y_train
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier 
```

```python
LR = LogisticRegression()
LR.fit(X_train,y_train)

KNN = KNeighborsClassifier()
KNN.fit(X_train,y_train)

SVC = SVC()
SVC.fit(X_train,y_train)

DTC = DecisionTreeClassifier()
DTC.fit(X_train,y_train)

RFC = RandomForestClassifier()
RFC.fit(X_train,y_train)

GBC = GradientBoostingClassifier()
GBC.fit(X_train,y_train)
```

# Predicting on Test data

```python
y_pred1 = LR.predict(X_test)
y_pred2 = KNN.predict(X_test)
y_pred3 = SVC.predict(X_test)
y_pred4 = DTC.predict(X_test)
y_pred5 = RFC.predict(X_test)
y_pred6 = GBC.predict(X_test)
```

# Evaluating the Algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
print('ACC LR', accuracy_score(y_test,y_pred1))
print('ACC KNN', accuracy_score(y_test,y_pred2))
print('ACC SVC', accuracy_score(y_test,y_pred3))
print('ACC DTC', accuracy_score(y_test,y_pred4))
print('ACC RFC', accuracy_score(y_test,y_pred5))
print('ACC GBC', accuracy_score(y_test,y_pred5))
```

```python
Scores = pd.DataFrame({'Models':['LR','KNN','SVC','DTC','RFC','GBC'],
                       'Scores':[accuracy_score(y_test,y_pred1)*100,
                                  accuracy_score(y_test,y_pred2)*100,
                                  accuracy_score(y_test,y_pred3)*100,
                                  accuracy_score(y_test,y_pred4)*100,
                                  accuracy_score(y_test,y_pred5)*100,
                                  accuracy_score(y_test,y_pred5)*100]})
```
